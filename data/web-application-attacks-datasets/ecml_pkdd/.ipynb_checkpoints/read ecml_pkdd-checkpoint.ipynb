```python
def parse_request(request):
    lines = request.split('\n')
    parsed_lines = [line.split(': ', maxsplit=1) for line in lines]
    return parsed_lines[:-1]
```

```python
def create_dataset(filename):
    samples = pd.DataFrame(columns=['id', 'class', 'vector'])
    i = 0
    for line in open(filename, 'r'):
        # Remove '\n' at the end of line
        line = line.rstrip()
        line = line.split(': ', maxsplit=1)

        if line[0] == 'Start - Id':
            lines = []
            sample = {'id': i}

        elif line[0] == 'End - Id':
            sample['class'] = lines[1][1]
            request_line = lines[2][0].split()
            if request_line[0] == 'GET':
                sample['vector'] = request_line[1]
            elif request_line[0] == 'POST' or request_line[0] == 'PUT':
                sample['vector'] = lines[-2][0]
            i += 1
            samples = samples.append([sample])

        lines.append(line)
        
        if i %% 1000:
            print(f'{i} requests processed')

    return samples.reset_index(drop=True)
```

```python
from http.server import BaseHTTPRequestHandler
from io import BytesIO

class HTTPRequest(BaseHTTPRequestHandler):
    def __init__(self, request_text):
        self.rfile = BytesIO(request_text.encode('utf-8'))
        self.raw_requestline = self.rfile.readline()
        self.error_code = self.error_message = None
        self.parse_request()

    def send_error(self, code, message):
        self.error_code = code
        self.error_message = message
```
